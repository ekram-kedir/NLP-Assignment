```python
from collections import Counter

# Remove symbols from the word
def remove_symbols(word):
    return ''.join(char for char in word if (('\u1200' <= char <= '\u137C') and (char != '\u1362') and (char != '\u1363') and (char != '\u1366')and (char != '\u1367')and (char != '\u2018')and (char != '\u2019') and (char != '\u1364')) or char.isspace())

# Remove non-Amharic-alphabetic characters and symbols from words
def clean_text(line):
    words = line.split() if len(line) > 1 else []
    cleaned_words = [word for word in words if remove_symbols(word)]
    return ' '.join(cleaned_words)

#Calculate the probability of the given sentence
def calculate_sentence_probability(file_path, given_sentence, batch_size=10000, encoding='utf-8'):
    sentence_counts = Counter()
    given_sentence = ' '.join(remove_symbols(word) for word in given_sentence.split())

    with open(file_path, 'r', encoding=encoding) as file:
        while True:
            lines = file.readlines(batch_size)
            if not lines:
                break
            for line in lines:
                cleaned_line = clean_text(line)
                # Check if the target sentence exists in the cleaned line
                if given_sentence in cleaned_line:
                    sentence_counts.update([given_sentence])

    total_sentences = sum(sentence_counts.values())

    # Calculate probability of the target sentence
    sentence_probability = sentence_counts[given_sentence] / total_sentences if total_sentences > 0 else 0

    return sentence_probability

# Specify the path for our file
file_path = '/kaggle/input/nlp-assignment/GPAC.txt'

# Specify the given sentence
given_sentence = "ኢትዮጵያ ታሪካዊ ሀገር ናት"

# Calculate the probability of the given sentence
probability = calculate_sentence_probability(file_path, given_sentence)
print(f'Probability of sentence "{given_sentence}": {probability:.6f}')
```
