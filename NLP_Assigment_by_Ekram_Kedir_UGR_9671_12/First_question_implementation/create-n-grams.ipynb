```python
# Remove symbols from the word
def remove_symbols(word):
    return ''.join(char for char in word if (('\u1200' <= char <= '\u137C') and (char != '\u1362') and (char != '\u1363') and (char != '\u1366')and (char != '\u1367')and (char != '\u2018')and (char != '\u2019') and (char != '\u1364')) or char.isspace())

# Remove non-Amharic-alphabetic characters and symbols from words
def clean_text(line):
    words = line.split() if len(line) > 1 else []
    cleaned_words = [word for word in words if remove_symbols(word)]
    return ' '.join(cleaned_words)

#Create the model
def ngrams_model(file_path, n, batch_size=1000, encoding='utf-8'):
    ngrams = []

    with open(file_path, 'r', encoding=encoding) as file:
        while True:
            lines = file.readlines(batch_size)
            if not lines:
                break
            for line in lines:
                # Cleaning each line
                cleaned_line = clean_text(line)
                words = cleaned_line.strip().split()
                # Generate n-grams
                ngrams.extend(zip(*[words[i:] for i in range(n)]))

    return [' '.join(ngram) for ngram in ngrams[:10]]

# Specify the path for our file
file_path = '/kaggle/input/nlp-assignment/GPAC.txt'

# Print each 10 ngrams for the language model
for n in range(1, 5):
    print(f'\n{n}-grams:')
    ngrams = ngrams_model(file_path, n)
    for ngram in ngrams:
        print(ngram)
```
