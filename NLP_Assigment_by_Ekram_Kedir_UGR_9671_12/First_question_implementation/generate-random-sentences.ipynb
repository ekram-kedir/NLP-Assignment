```python
import random
from collections import Counter

# Remove symbols from the word
def remove_symbols(word):
    return ''.join(char for char in word if (('\u1200' <= char <= '\u137C') and (char != '\u1362') and (char != '\u1363') and (char != '\u1366')and (char != '\u1367')and (char != '\u2018')and (char != '\u2019') and (char != '\u1364')) or char.isspace())

# Remove non-Amharic-alphabetic characters and symbols from words
def clean_text(line):
    words = line.split() if len(line) > 1 else []
    cleaned_words = [word for word in words if remove_symbols(word)]
    return ' '.join(cleaned_words)

#Calculate the probability first so thst we can get the sentences with high probability
def calculate_ngram_probabilities(file_path, n, top_k, batch_size=10000, encoding='utf-8'):
    ngram_counts = Counter()

    with open(file_path, 'r', encoding=encoding) as file:
        while True:
            lines = file.readlines(batch_size)
            if not lines:
                break
            for line in lines:
                cleaned_line = clean_text(line)
                words = cleaned_line.strip().split()
                ngrams = [' '.join(words[i:i+n]) for i in range(len(words)-n+1)]
                ngram_counts.update(ngrams)

    # Sort n-grams by frequency and take the top_k
    top_ngrams = ngram_counts.most_common(top_k)

    total_ngrams = sum(ngram_counts.values())

    # Calculate probabilities for the top_k n-grams
    ngram_probabilities = {ngram: count / total_ngrams for ngram, count in top_ngrams}

    return ngram_probabilities

#This function will generate a sentence
def generate_random_sentence(ngram_probabilities, n):
    sentence = []
    
    # Choose a starting word randomly
    starting_word = random.choice(list(ngram_probabilities.keys()))
    sentence.extend(starting_word.split())

    # Generate the rest of the sentence using n-grams
    for _ in range(n - 1):
        current_ngram = ' '.join(sentence[-n+1:])
        next_word_candidates = [word.split()[-1] for word in ngram_probabilities.keys() if word.startswith(current_ngram)]
        if next_word_candidates:
            next_word = random.choice(next_word_candidates)
            sentence.append(next_word)
        else:
            break  # Stop if no more valid n-grams

    return ' '.join(sentence)

# Specify the path for our file
file_path = '/kaggle/input/nlp-assignment/GPAC.txt'

# Choose an n value
n_value = 3

# Calculate n-gram probabilities for the specified n
ngram_probabilities = calculate_ngram_probabilities(file_path, n_value, 100)

# Generate and print five random sentences with a length of n words
for _ in range(5):
    random_sentence = generate_random_sentence(ngram_probabilities, n_value)
    print(random_sentence)
```
